```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# linspace(start end , number of values)
x = np.linspace(4,15,100)
y = np.sin(x)
```

## Line Plot

```python
x
```

```python
y
```

```python
plt.figure(figsize= (5,3))
plt.plot(x, y, label = "Sin Wave", linestyle ="dotted", linewidth =2, color ="#000000")  # color excepts hex code or name of color
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin Wave")
plt.legend()
plt.grid(True)
plt.show()
```

## Bar Chart

```python
category = ["A","B","C", "D","E"]
value = [33, 54, 42, 51, 29]
```

```python
plt.figure(figsize= (5,3))
plt.bar(category, value,alpha = 0.5, edgecolor = "black", color = ["#256789", "#234781", "#234290", "#278910", "#287658"])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparing Catogories")
plt.grid(True)
plt.show()
```

## Scatter Plot

```python
x = np.random.rand(100)
y = np.random.rand(100)
color = np.random.rand(100)
size = 1000 * np.random.rand(100)
```

```python
plt.figure(figsize= (5,3))

# cmap is color map use for creting gradient, ex = plasma, cividis, viridis, blues
plt.scatter(x, y, c = color, s = size, alpha = 0.5, cmap = "plasma")  
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.colorbar(label = "Color Intensity")
plt.grid(True)
plt.show()
```

## Histogram

```python
dice = np.array([1,2,3,4,5,6])
sample = np.random.choice(dice, 4) # Rolling dice 4 times in a row and getting values
print("Output of 4 samples", sample)
print("Mean of above samples", sample.mean())
```

```python
samples_1000 = [np.random.choice(dice, 4).mean() for i in range(1000)]
samples_1000
```

```python
plt.figure(figsize = (5,3))
plt.hist(samples_1000, alpha = 0.5, edgecolor = "black", bins = 50, color = "red", histtype = "barstacked")
plt.show()
```

```python
data = np.random.randn(10000)
plt.figure(figsize = (5,3))
plt.hist(data, alpha = 0.4, edgecolor = "black", color = "#278910", bins = 50, histtype = "barstacked")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True)
plt.show()
```

## Pie Chart

```python
sizes = np.random.randint(35, 89, 4)
labels = ["East", "West", "North", "South"]
plt.figure(figsize = (5,4))
plt.pie(sizes, labels = labels, colors = ["red","lightcoral","cyan","yellow"],autopct = "%1.3f%%", explode = [0,0.1,0,0])
plt.title("Pie Chart")
plt.show()
```

## Box Plot

```python
data = np.random.normal(loc = 0, scale = 1, size= (100,))
plt.figure(figsize=(5,4))
plt.boxplot(data, patch_artist = True, vert = False)
plt.title("Box Plot")
plt.show()
```

## Polar Plot

```python
angles = np.linspace(0, 2* np.pi, 100)
radii = np.random.rand(100)
```

```python
plt.figure(figsize = (5,4))
ax = plt.subplot(111, polar =True)  # subplot = 111 = 1 row , 1 column , 1 is sub plot
ax.plot(angles, radii, color = "red", linewidth = 0.2)
ax.fill(angles, radii, color="blue", alpha = 0.4)
plt.title("Polar Plot")
plt.show()
```

## HeatMap
