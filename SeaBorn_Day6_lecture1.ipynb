```python
import seaborn as sns
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
```

```python
height = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]
weight = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
```

```python
%matplotlib inline
plt.figure(figsize = (3,2))
plt.xlabel("Height")
plt.ylabel("Width")
plt.title("Height & Width")
sns.scatterplot(x = height, y= weight)
# plt.show()
```

## Count Plot

```python
gender =["Female","Female","Female","Female","Male","Male","Male","Male","Male","Male"]
```

```python
plt.figure(figsize = (4.5,3))
plt.title("Count of Male and Female")
sns.countplot(x=gender, hue=gender, legend = ["Female","Male"], fill = False) # changes the graph direction , hue is use to give different colors to the graph
plt.show()
```

## Working with Datafromes with countplot()

```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("SampleSuperstore.xlsx")
df.head()
```

```python
plt.figure(figsize=(6.5,3))
sns.countplot(data=df, x="Ship Mode", hue="Segment")
plt.show()
```

## Scatter Plot

```python
df.columns
```

```python
# Scatter plot of column sales and quantity
plt.figure(figsize=(4.5, 3))
sns.scatterplot(data = df, x = 'Quantity' , y='Sales', hue = "Category")
plt.show()
```

## HistPlot

```python
plt.figure(figsize=(20, 6))
sns.histplot(data = df, x = "Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(4.5, 3))
sns.scatterplot(data=df, x="Quantity", y="Sales", hue="Category", hue_order=["Technology", "Office Supplies", "Furniture"],
                palette={"Technology":"red", "Office Supplies":"#234567",
                        "Furniture":"green"})
plt.show()
```
